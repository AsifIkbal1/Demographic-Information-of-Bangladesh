# **Import data (City Corporation)**

```python
gpkg_path = '/kaggle/input/bangladesh-census-city-corporation-data-2022/citycorporation.gpkg'
gdf = gpd.read_file(gpkg_path)
```

# > **Show data list**

```python
gdf.head() # List all input data
```

> # **Visualize the change rate of values**

```python
df = gdf.drop(columns='geometry')
plt.figure(figsize=(50,6))# Adjust the figure size if needed
df.plot.line()
plt.title('Line Plot of DataFrame')
plt.xlabel('Index')
plt.ylabel('Values')
plt.show()
```
